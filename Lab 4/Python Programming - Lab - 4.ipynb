### 06) WAP to print sum of series 1 + 4 + 9 + 16 + 25 + 36 + ...n.

```python
print("Sum of squares series:")
n = int(input("Enter value of n: "))
sum_squares = 0
for i in range(1, n + 1):
    sum_squares += i ** 2
print(f"Sum of series 1² + 2² + ... + {n}² = {sum_squares}")
```

### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
print("Sum of alternating series:")
n = int(input("Enter value of n: "))
alt_sum = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        alt_sum -= i 
    else:
        alt_sum += i 
print(f"Sum of series 1-2+3-4+...+{n} = {alt_sum}")
```

### 10) WAP to print GCD of given two numbers.

```python
print("GCD of two numbers:")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

a, b = num1, num2
while b != 0:
    a, b = b, a % b
print(f"GCD of {num1} and {num2} = {a}")
```

### 12) WAP to find whether the given number is Prime or not.

```python
print("Prime number check:")
num = int(input("Enter a number: "))
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

### 13) WAP to print sum of digits of given number.

```python
print("Sum of digits:")
num = int(input("Enter a number: "))
temp = abs(num)
digit_sum = 0

while temp > 0:
    digit = temp % 10
    digit_sum += digit
    temp //= 10

print(f"Sum of digits of {num} = {digit_sum}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
print("Palindrome check:")
num = int(input("Enter a number: "))
temp = num
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp //= 10

if num == reverse:
    print(f"{num} is a palindrome")
else:
    print(f"{num} is not a palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
print("Armstrong number check:")
num = int(input("Enter a number: "))
temp = num
power = len(str(num))
armstrong_sum = 0

while temp > 0:
    digit = temp % 10
    armstrong_sum += digit ** power
    temp //= 10

if num == armstrong_sum:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
print("Perfect numbers between 1 to n:")
n = int(input("Enter value of n: "))
print(f"Perfect numbers between 1 and {n}:")

for num in range(1, n + 1):
    divisors_sum = 0
    for i in range(1, num):
        if num % i == 0:
            divisors_sum += i
    
    if divisors_sum == num:
        print(num, end=" ")

print("\n" + "="*50)
```
