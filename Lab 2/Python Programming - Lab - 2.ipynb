### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
d = float(input("Enter a degree : "))
print(f"Fahrenheit : {(d + 9 / 5) + 32}")
f = float(input("Enter a Fahrenheit : "))
print(f"Fahrenheit : {(f - 32) * 5 / 9}")
```
